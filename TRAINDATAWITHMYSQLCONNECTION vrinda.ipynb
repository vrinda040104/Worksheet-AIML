# WHAT IS UNIVARIATE ANALYSIS

# HOW MANY PEOPLE SURVIVED AND HOW MANY DIED PLOT ON GRAPH

```python
import pandas as pd

print("SURVIVED:",len(df_mysql[df_mysql["Survived"]==1]))
print("DIED:",len(df_mysql[df_mysql["Survived"]==0]))


Survived=len(df_mysql[df_mysql["Survived"]==1])
Died=len(df_mysql[df_mysql["Survived"]==0])


import matplotlib.pyplot as plt

labels = ['Survived','Died']
sizes = [Survived,Died]
colors = ['lightgreen', 'pink']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')
plt.title('SURVIVED AND DIED PEOPLE GRAPH')
plt.show()
```

# HOW MANY PASSENGERS WERE IN FIRST CLASS,SECOND CLASS AND THIRD CLASS PLOT THOSE FIGURES ON GRAPH

```python
import pandas as pd

print("First Class:",len(df_mysql[df_mysql["Pclass"]==1]))
print("Second Class:",len(df_mysql[df_mysql["Pclass"]==2]))
print("Third Class:",len(df_mysql[df_mysql["Pclass"]==3]))

First_Class=len(df_mysql[df_mysql["Pclass"]==1])
Second_Class=len(df_mysql[df_mysql["Pclass"]==2])
Third_Class=len(df_mysql[df_mysql["Pclass"]==3])

import matplotlib.pyplot as plt

labels = ['First Class', 'Second Class', 'Third Class']
sizes = [ First_Class,Second_Class,Third_Class]
colors = ['yellow', 'pink', 'lightblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Passenger Distribution by Class')
plt.show()

```

# HOW HAS BETTER CANCE OF SURVIVAL MALE OR FEMALE

```python
import pandas as pd
import matplotlib.pyplot as plt

survival_counts = df_mysql.groupby(['Sex', 'Survived']).size().unstack()
fig, ax = plt.subplots()
ax.pie(survival_counts[1], labels=survival_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'pink'])
ax.set_title('Survival Distribution by Gender')
plt.show()
```
